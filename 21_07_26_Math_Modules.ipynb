```python
##Create a scientific calculator that supports square root, factorial, logarithm, power, trigonometric functions, GCD, and LCM
import math

print("===== SCIENTIFIC CALCULATOR =====")
print("1. Square Root")
print("2. Factorial")
print("3. Logarithm")
print("4. Power")
print("5. Trigonometric Functions")
print("6. GCD")
print("7. LCM")

choice = int(input("Enter your choice: "))

if choice == 1:
    number = float(input("Enter a number: "))
    print("Square Root =", math.sqrt(number))

elif choice == 2:
    number = int(input("Enter a number: "))
    print("Factorial =", math.factorial(number))

elif choice == 3:
    number = float(input("Enter a number: "))
    print("Logarithm =", math.log10(number))

elif choice == 4:
    base = float(input("Enter base: "))
    exponent = float(input("Enter exponent: "))
    print("Power =", math.pow(base, exponent))

elif choice == 5:
    angle = float(input("Enter angle in degrees: "))
    rad = math.radians(angle)

    print("Sin =", math.sin(rad))
    print("Cos =", math.cos(rad))
    print("Tan =", math.tan(rad))

elif choice == 6:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("GCD =", math.gcd(a,b))

elif choice == 7:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("LCM =", math.lcm(a,b))

else:
    print("Invalid Choice")
```
